###### 2. Types of Services
***

Introduce the three basic categories of cloud services: i. SaaS, or Software as a Service, ii. PaaS, or Platform as a Service and iii. IaaS, or Infrastructure as a Service. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesSoftwareasaServiceSaaS"></a>
***
###### Software as a Service - SaaS
***

SaaS, this is a software distribution model whereby you have a third-party provider of some kind and they effectively host the application for you. 

That application is then made available to customers over the Internet.

Now we typically aren't talking about desktop-level applications such as word processors or spreadsheets but rather larger scale, multi-user applications such as enterprise resource planning, or customer management software or maybe anything that is driven by a database. 

So again, the idea is that everything is available as a service. 

The reason why you see that aaS after all of them is because there is a lot of infrastructure required to provide these services. 

So the basic idea is that, rather than you investing in that infrastructure yourself, you simply subscribe to those who already have it. 

If you needed to implement a large scale CRM application in your own environment - would of course have to invest in the servers, the storage, the networking, the software itself, the licensing, then the deployment, the configuration, the management, the upkeep. 

All of that has to be in place for you to effectively use that application. 

That's all well and good, but again these providers already have all of that. 

So you simply subscribe to it, you get a portion if you will, or a section of that application that you can use and now you don't have to invest in any of that infrastructure, it's already there and again, when it comes to cloud services, this is applicable to all cloud services.

You only pay for what you use. So if you were a smaller environment, and you needed to support let's say only maybe a few dozen users in a CRM application. 

Then you would only pay for that amount, if you were a much larger organization, then you had to support hundreds of users, of course you would pay more. 

But in both cases, you still don't have to invest in the infrastructure. 

It's already there, so you just subscribe to it. Everything is effectively a pay-as-you-go model. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesSoftwareasaServiceSaaSAdvantages"></a>
***
###### Advantages
***

So in terms of advantages, the time to deployment is very rapid. 

You don't have to acquire all of that hardware, and software, and licensing. 

It's already there, so you really can get started almost immediately, greatly easing the administration as well because, again, all of that infrastructure that needed to be implemented in the on-premise environment is just gone, so you don't have to look after the servers themselves, or the network configuration or anything along the lines of licensing, all of that goes away. 

It is highly scalable because, again, the providers have tremendous amounts of resources available. 

It is very accessible, it only requires Internet access and a browser and again, because of the degree of robustness, if you will, at the provider, it is very resilient. 

They have redundancy built in to just about every aspect of every service, so it's very fault-tolerant.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesSoftwareasaServiceSaaSDisadvantages"></a>
***
###### Disadvantages
***

And of course, there are disadvantages. 

Security, the privacy of sensitive information, and the compliance, the location of the data are typically foremost in the concerns of users. 

Quite simply, the data itself is not under your physical control and really you don't even know where it actually resides. 

Plus, in most cases, you are sharing the storage of that data, so if I do subscribe to my CRM application, my little portion of it is kept separate from any other given client. 

But it is still possibly available to the people who manage the environment itself where the data actually resides, so the privacy is a concern, and again, the fact that it is physically not under your control can be quite limiting. 

Performance can also be an issue because you are ultimately using the Internet to access that application, so that of course can cause problems.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesPlatformasaServicePaaS"></a>
***
###### Platform as a Service - PaaS
***

Now, PaaS or Platform as a Service is a model used typically to develop and deliver applications over the Internet. 

It delivers software tools in development environments to users, again, as a service and this really is the same type of scenario whereby if you want to create your own development environment in-house, you still need the servers. 

You still need the networking configuration, you still need the software for the development environment, so all of that can be made available within your cloud subscription as well and then ultimately you do not have to invest in all of that infrastructure. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesPlatformasaServicePaaSAdvantages"></a>
***
###### Advantages
***

Again, there are tremendous advantages. 

For testing and deployment, you can easily test different configurations. 

You can create virtual machines in any configuration that you want to ensure that your application will run across multiple different operating systems for example and you can very quickly add new features to an application. 

So this dynamic allocation is very easy to implement, there is improved support and minimal management - again, because there is just so much less in terms of the infrastructure within your own environment.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesPlatformasaServicePaaSDisadvantages"></a>
***
###### Disadvantages
***

Again, there are disadvantages, that lack of control is always a concern. It is typically a shared platform, all of the other subscribers are using the same ultimate environment as you are. 

So data security still remains a concern when it comes to the privacy and the retention of data. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesInfrastructureasaServiceIaaS"></a>
***
###### Infrastructure as a Service - IaaS
***

And then finally, Infrastructure as a Service, or IaaS, provides things like virtual machines. 

The operating system itself, the networking capabilities, and the security, so getting close to the hardware level, but this is the lowest level of service. 

So if you need more servers in your environment for example, you can just create new virtual machines - they can come with any operating system that you want in many cases. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesInfrastructureasaServiceIaaSAdvantages"></a>
***
###### Advantages
***

Advantages here, cost savings, again, that lower infrastructure cost. 

You don't need the physical systems to host the virtual machines. 

They are already there at the provider, again, highly scalable. 

You can quickly scale up or down, depending on the requirements, and highly available as well. 

It is really up to the provider to ensure that the redundancy is there for all of these features. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConceptsTypesofServicesInfrastructureasaServiceIaaSDisadvantages"></a>
***
###### Disadvantages
***

Again, disadvantages, maintenance may not be included. 

Now that depends on how you approach this if you will, but with respect to things like versioning and upgrades, that is usually up to you. 

But if we just think about a virtual machine, for example. 

All you see is the virtual machine, you can manage it however you want, so let's imagine you create a virtual machine with a particular version of Windows Server. 

The versioning, and the upgrades, and the updates for that server are up to you to manage. 

But the underlying hardware is up to the provider, so that's why I say maintenance may not be included, it depends on what you are talking about. 

But keeping the hardware current, and upgraded, and up to date, that is absolutely up to the provider. 

Again, security may not be adequate depending on the requirements and there are always legal considerations typically depending on where data actually resides. 

But ultimately, you can subscribe to any or all of these levels of service, depending on what you need. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModels"></a>
***
###### 3. Cloud Delivery Models
***

In this presentation we'll take a look at several different types of clouds known as a private cloud, a public cloud, a hybrid cloud, and a community cloud. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPrivate"></a>
***
###### Private
***

Uses a proprietary architecture that is dedicated to a single organization, hence the term private. 

It can be ideal for businesses with dynamic or unpredictable computing needs or just businesses that require more direct control over the configuration. 

The idea typically is that you would have a fairly large organization that already has, let's say, a corporate headquarters or a fairly robust data center in place and then multiple smaller branch offices. So rather than investing in separate and independent infrastructure configurations for each of those branch offices, they will simply subscribe to the infrastructure available in the corporate headquarters. 

So it looks just like a cloud configuration from any given user's perspective, in that there really is no infrastructure in the branch office but it's all under your control, so as a single organization, you manage and maintain and implement everything, hence the term, again, private cloud. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPrivateAdvantages"></a>
***
###### Advantages
***

So the advantages here, certainly security. Everything remains under your control. 

So that has traditionally been a bit of a hurdle with respect to public cloud services, because there is an awful lot that is not under your control. 

Performance, you could have improved transfer rates, because you can use dedicated lines and connections and of course, management, things such as customized hardware can be implemented where that might not be an option in a public cloud configuration.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPrivateDisadvantages"></a>
***
###### Disadvantages
***

Now the disadvantages, certainly higher costs. 

You do have to make that investment in the hardware, and the personnel at least at the corporate headquarters but a lot of organizations already have that in place. 

But there is still the ongoing maintenance at the data center - there can be very high demands with respect to power, cooling, and damage of course to power surges, or any kind of weather damage, or any sort of natural disaster. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPublic"></a>
***
###### Public
***

So then, the public cloud is a platform in which service providers are used to offer their infrastructure, their resources, their applications, their storage and a tremendous number of other services but these are made publicly available to any one, so we have any number of customers subscribing to those services.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPublicAdvantages"></a>
***
###### Advantages
***

Advantages here, certainly cost savings. 

The organizations are not required to purchase, install, or maintain servers, or any other equipment and it's also what's known as a multi-tenant environment, meaning there are multiple customers that are all sharing the computing resources. 

Now you might think, well, that's a bit of a disadvantage, but most of the large scale providers have tremendous amounts of physical resources available.

So with respect to what you need, in most implementations, it really can be considered to be almost unlimited, so really that's not so much of a concern. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsPublicDisadvantages"></a>
***
###### Disadvantages
***

In terms of disadvantages however, there is limited control over privacy and security and again, that remains a stumbling block for a lot of organizations. 

It's just not recommended for sensitive applications or information. 

The data does not reside under your physical control. 

There can also be limited flexibility - the platform depends on the platform provider. 

Now, that said, these days most of the large scale providers do offer an awful lot of flexibility but there can still be some limitations whereby provider X simply does not offer the service you need. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsHybrid"></a>
***
###### Hybrid
***

Then there is the hybrid cloud, which is a combination of both public and private clouds. 

This helps to maximize the efficiency and the performance and it's certainly gaining in popularity, as it offers the benefits of both the public and the private models. 

So if you already have a private cloud, and you are perhaps looking to extend, well, you can consider the public cloud as the extension, if you will, you can see that you can bridge the two, and implement workloads that quite literally span both. 

So that if you need to offer very private services, within your own organization, that can be done in the private portion. 

But if you need to make various workloads and services available to the public Internet, that can be done in the public portion. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsHybridAdvantages"></a>
***
###### Advantages
***

So it does have similar advantages to just a public cloud. 

It is very scalable. 

Again, the providers have effectively unlimited resources, so when you do decide to extend, those resources are there. 

It can be more cost efficient because, again, you aren't expanding your own environment, so you don't have to invest in that additional hardware. 

Better overall security and greater flexibility. Basically, the fact that you can keep whatever it is that you need to keep private on the private side, then anything that needs to be made publically available can be offered in the public side. 

So you have that flexibility and enhanced security, because you have better control over which data resides where.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsHybridDisadvantages"></a>
***
###### Disadvantages
***

But again, there are always disadvantages. 

There can be an infrastructure dependency. 

There might be some compatibility issues with what you have internally versus what you need externally and again, there can always be some security concerns, possible security breaches when utilizing the public cloud side, for example. 

There might be some kind of a avenue into the private from the public, using the connection you have established. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsCommunity"></a>
***
###### Community
***

The community cloud is more of a collaborative effort wherein infrastructure is shared between several organizations.

So what you see is perhaps a partnership of various internal private organizations just all getting together, if you will, to share these services. 

So there are still concerns - can still be security issues, compliance issues, and jurisdiction issues. 

In other words, who looks after this aspect of that service? So these aren't nearly as common but you can still see these types of implementations and they can be a combination of private and public as well. 

So you may see some hybrid configuration within the community. 

So there can be a fair amount of overlap, but in essence, this is typically some kind of cloud configuration between a collection of organizations but again, they certainly aren't as common as the private, or the public, or even the hybrid. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModels"></a>
***
###### 3. Cloud Delivery Models
***

Now, if you are using cloud services, this introduces the question of how you will connect from within your own on-premises organization to the cloud provider, so, in this presentation, we'll take a look at a couple of options for doing so. 

But before getting to that, I would just like to address the term connectivity and what it means in this context.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsConnectivityMethods"></a>
***
###### Connectivity Methods
***

So let's imagine that you create your cloud subscription and you decide that you need to host a website or a web-based application, of course, you can do so. 

So you might create a virtual machine to host the site or the application. And in terms of your customers, your business partners, and even your own users, accessing the site or the application, they will just use a browser, standard Internet connectivity. 

So that's all well and good, but that's not really what connectivity refers to in this context, if you are the administrator and/or the developer of that site or application then you need connectivity to the server. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsConnectivityMethodsVPN"></a>
***
###### VPN
***

And lets scale this up, imagine you have possibly dozens upon dozens of servers. 

Then you need those servers available to you, and you need them to behave, more or less, as if they were in your network, so, to accomplish that kind of configuration, a VPN is certainly a viable option, and perhaps one of the simpler options for connecting to the cloud provider. 

Typically, the cloud services offer VPN appliances or native software VPNs that you can use to ensure that all of those servers in the cloud are available to you on your network, so that you can manage their resources, perform maintenance and updates and security patches, things along those lines. 

Now, of course, there are costs associated with this but like all things cloud related, you only pay for what you use. 

But the cloud provider may charge for the total bandwidth transferred over that VPN. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts3CloudDeliveryModelsConnectivityMethodsDirectConnection"></a>
***
###### Direct Connection
***

Now another option is direct connection, and this can be implemented in a few different flavors, if you will. 

But this offers a more predictable connection. 

Typically, you'll see cloud providers using co-location facilities to offer direct private connections and, often, those co-location facilities are telecom providers. 

People who already have communications infrastructure in place. 

So customers will then, in turn, rent these facilities from the data centers and they will, in turn, offer a cross connect back to the public cloud. 

So what you'll often find is that any given cloud service provider will offer this using this telecom provider, or that one, or this one, or that one and you can kind of browse through and find one that is appropriate for your area. 

But a lot of those telecom providers have these dedicated connections back to the public cloud providers, so, in essence, they are reselling that service.

Now in this kind of configuration, of course, there are costs.

The data center will charge for the cross connect, itself and the cloud provider may still charge for the port. 

ut once you have this, most providers will allow unlimited transfers over the direct connect. 

So this lowers the per-gigabyte transfer cost. 

Now there may be commitments involved, as well, but the terms can be as short as one month, so generally, you don't have to lock into a multi-year contract or anything along those lines. So the immediate upfront cost might be higher, but given the fact that they generally offer unlimited transfers, this is where you start to see the savings because if you're paying a per-gigabyte fee per month, then sooner or later, that will probably cost more than this type of configuration.

But, of course, it always depends on the scale, how much data is going to actually cross that connection? 

Should also mention, if it is a fairly small scale implementation, you may not need either of these. 

If, for example, you only have a few servers in the cloud, hosting a few applications or services, and you need to connect to them, you certainly can using something like remote desktop. 

You can simply connect to any given server and remotely administer it like any other one - it just might not give you very good access to the resources on that server, but it is generally fine for just administrative tasks. 

If you just need to change some configuration options, or implement an update or a patch, or things along those lines, then that is still just regular Internet connectivity. 

So like anything, as mentioned, it depends on the scale but those are some of the methods available when you do need that more reliable connectivity to your cloud-based resources.

<a id="cloudConcepts5CloudSecurity"></a>
***
###### 5. Cloud Security
***

While cloud services can provide a number of benefits, it's not the kind of thing where you typically see organizations saying all right, great, let's just immediately move everything to the cloud. 

It's usually a slow process and there has to be a lot of investigation, perhaps proof of concept has to be demonstrated, things along those lines, so in other words, before you do make a move to the cloud, you need to understand what's involved and in most cases that is knowing the risks. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactors"></a>
***
###### Risk Factors
***

So in this presentation we'll take a look at several of the risk factors at various levels and for various types of cloud implementations, beginning with IaaS or Infrastructure as a Service.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsIaaS"></a>
***
###### IaaS
***

- With this you might have issues with legal transborder requirements. When you create any kind of cloud subscription, you really do not know where the data center is, or perhaps more specifically, where your data resides. And most of the larger providers tend to implement replication schemes to protect your data so that it is actually in multiple data centers. But that might actually involve that data crossing a border of some degree. So that is something you need to be aware of and whether or not that is acceptable by your organization, and/or perhaps data privacy laws. 

- Failure of course is always an issue, how isolated are you when it comes to any given failure? Because it is a multi-tenant environment, if this fails for that tenant, will it affect me? So you need to know what kind of isolation is in place. 

- Backup and security measures, you certainly need to know what is being implemented there. Do they have intrusion detection systems and intrusion prevention systems? What are their disaster recovery procedures and/or policies for making backups? And I'm not trying to suggest at all that these things aren't in place. In fact for most of the large scale providers, they are very much so in place. It's just a matter of knowing what they are. 

- Same thing with the physical security, what kind of physical security do they offer at their data centers, and how do they handle data disposal? 

So again these are all just things to consider. It's not that they aren't there, it's just that you need to know what they are and how these risks are mitigated.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsPaaS"></a>
***
###### PaaS
***

So then at the PaaS level, or platform as a service.

- There can be issues with application mapping, theft and disclosure might occur. 

- There can be SOA related vulnerabilities, that's service-oriented architecture. You don't really know how the service is structured at the provider end, if you will. So there can be vulnerabilities there that you don't know about.

- Application disposal, what happens to the original? So as your versions increase and you move from ultimately testing to production. What happens to the original files and/or any backups that were made?

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsSaaS"></a>
***
###### SaaS
***

And then at the Software as a Service level there can be questions about data ownership.

- Again, the data is not physically under your control. It resides who knows where, so there could be issues with ownership.

- And again disposal becomes an issue, you need to ensure in many cases that it is properly disposed, and that no one else can get a hold of your data. 

- Software systems development life cycle, what happens with respect again through the whole versioning history of applications? 

- Identity and access management, all of this has to be considered. Who looks after that kind of user management, for example. 

- Broad exposure, there can be issues when it comes to any given application and/or the data, and/or the users, with respect to who is able to see what. 

- There can be browser vulnerabilities, because in most cases you are using a browser to access the application. 

- And ultimately, you need to understand what kind of exit strategy can exist if/and/or when you are going to discontinue the service. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsCloud"></a>
***
###### Cloud
***

Then with respect to the different types of clouds. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsCloudPublic"></a>
***
###### Public
***

- When dealing with a public cloud, there can be issues with data pooling. The infrastructure is shared by multiple tenants and tenants have no relations to other tenants or to any other enterprise. So there's no common concerns for the security of the data itself. In other words, my concerns aren't your concerns, and they aren't anyone else's concerns, so I am only concerned with my own interests - you are only concerned with your own interests. But those may not be compatible concerns. So how does the provider handle, in essence, trying to make everyone happy? 

- Collateral damage, an attack on one tenant could impact the other tenants - again, if a portion of a service goes down, how isolated are you from that failure? So that whole idea, again, that any kind of attack can propagate.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts5CloudSecurityRiskFactorsCloudPrivate"></a>
***
###### Private
***

- And if it's a private cloud, there can be issues with application compatibility - older applications may not work with newer technologies such as customized software. 

- There certainly is a significant investment. It requires a certain skillset, training, recruitment, and maintenance of the infrastructure is always necessary, so that is certainly a major consideration. Now at least with a private cloud things are more under your own control. But there are still these considerations.

###### Hybrid
***

- And then with the hybrid cloud, there can be issues, of course, with mixing two or more cloud types. This requires strict identity control and strong credential management.

- And still has risks such as unavailability of an application, or data loss, theft, disclosure - anything along those lines can still happen within a hybrid cloud configuration.

###### Community
***

- And finally, the community cloud, this involves sharing the cloud, so entities may have different security measures or requirements, even when belonging to the same enterprise. 

- So a faulty procedure or a service level agreement can render entities at risk. 

- And there are, again, possible impacts in the areas of data loss, theft, or disclosure. 

So ultimately there are always concerns and I do want to stress the major cloud providers will make heavy investments into mitigating these risks as much as possible. 

In essence, they depend on that ability, so it's not a matter of them effectively saying well, here is our service and here is how we can protect your interests.

***
###### 6. Local vs. Cloud
***

Now, in this presentation we'll take a look at some considerations for implementing storage both locally and in the cloud and like any comparison, there are always pros and cons, but beginning with local storage.

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts6LocalvsCloudLocal"></a>
***
###### Local
***

One of the main benefits is, of course, speed. 

Storing data locally is quite simply faster than uploading to the cloud, particularly if you are dealing with a lot of data on a regular basis. 

Even simple storage mechanisms, such as internal storage on a server is going to be faster than continually uploading. 

Of course, security and control is a major concern - when it's local, you have full control of backups - you have better overall control of who can access the data and in the case of external storage, you can disconnect the drives. 

You can keep the data safe from attacks - any kind of attack coming from the outside, regardless of the skill of the attacker, will not be able to reach a physically disconnected storage medium. 

So, you have those options, you can decide for yourself what kind of security you want to implement. 

Now, there are, of course, considerations with local storage. 

It can absolutely be tremendously expensive when you start getting up into the highest levels of SAN storage.

So you have to create and maintain some kind of local storage system. 

You need the hardware, you need the software to manage and maintain it, so that can be very expensive. 

You also have to consider on-site disasters which of course can destroy the data. 

So to mitigate that risk you typically have to implement off site storage which translates into more administrative overhead. 

So there are always considerations. There are always pros and cons. 

< [Table of Contents](#top) | [References](#references) >
<a id="cloudConcepts6LocalvsCloudCloud"></a>
***
###### Cloud
***

So if you then look at cloud storage, on the plus side, it is absolutely very cost effective. 

There is minimal maintenance because of all of that infrastructure, for example a SAN, is not required at all. 

Scalability is absolutely there - can increase or decrease the storage however you see fit, so it's not always just scaling up, in many cases you can scale back. 

Disaster recovery is effectively already addressed - can secure the data remotely and you know it will remain safe. 

You have that off-site storage simply by nature because it is at the cloud data center, plus, very likely, a local copy but even the cloud data centers themselves usually will replicate that data to other data centers. 

So you always have multiple copies and you certainly have the accessibility - can access the data from any device with an Internet connection. 

So typically the data that you tend to find in cloud storage is something like a database that drives in applications, so that when I open my application it doesn't matter where I am, it doesn't matter what I'm using. 

As long as I have an Internet connection, the application can access the data.

But, of course, there are the downsides. 

The speed, it requires a good Internet connection. 

For example if there is a lot of data to back up, it could absolutely take a long time without a good connection and perhaps the biggest issue is control. 

There is always sensitive data, so you need to know is it even acceptable for it to be stored anywhere other than within your own organization and if it is acceptable, then you need to understand what the risks are, and how you can mitigate them with respect to keeping that data sensitive. 

So, as mentioned, there are always considerations. 

There are always pros and cons, and it might depend, of course, on the application, what it is that you need to do, and where that data really is best suited to reside but you do have those options and in many cases it's possible that you can implement a mix of both. 

You might have local data that is entirely appropriate to be local and then you have data that should be in the cloud, so there's nothing wrong with that either - don't have to make the call where it's all or nothing.

It's dependent on what your needs are, and as mentioned, you can always decide for yourself where the data should best reside.
